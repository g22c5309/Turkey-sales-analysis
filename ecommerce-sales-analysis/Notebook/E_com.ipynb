## Data Exploration

Exploratory Data Analysis (EDA) is conducted to understand customer purchasing behavior, sales trends, and product performance. The analysis focuses on identifying patterns in revenue, customer activity, and product categories, as well as uncovering insights that could inform business decisions.

#### 1. Revenue Over Time

#### 2. Revenue by Product Category

This visualization shows the total revenue generated by each product category. The objective is to understand which categories contribute most to overall sales and to identify potential differences between high-revenue and low-revenue product groups.

<img src="Revenue by Product Category.png" width="500" height="700" />


Key Findings

* The strong performance of the Electronics category suggests high customer demand and potentially higher-priced items, while lower revenue in the Books category may reflect lower price points or reduced purchase frequency.



These results indicate that Electronics and Home and Garden are key revenue drivers for the business and may benefit from continued investment in marketing and inventory. Conversely, the Books category may require further analysis to determine whether pricing, promotion strategies, or product variety could be optimized to improve performance.

#### 3. Number of Orders by Product Category

This visualization shows the total number of orders placed for each product category. The purpose is to assess category popularity and compare order volume across different product groups.

<img src="Number of Orders by Product Category.png" width="500" height="700" />


Key Findings:

* Interestingly, although Electronics generates the highest revenue overall, it does not have the highest number of orders. This suggests that electronics purchases are less frequent but higher in value, whereas sports products are purchased more often but at lower average prices.

Overall, this analysis highlights the importance of examining both revenue and order volume. Categories with high order counts are key drivers of customer engagement, while high-revenue categories contribute more significantly to overall sales value.


##### 4. Revenue by Device Type

This dashboard compares total revenue and number of orders across different device types. The goal is to determine whether revenue differences are driven by purchase frequency or differences in average order value.

<img src="Dashboard 1.png" width="600" height="600" />

Key Finding:

* The analysis shows that mobile devices generate the highest total revenue and the highest number of orders compared to desktop and tablet users.

The results highlight the importance of mobile commerce in the business model. Optimizing the mobile shopping experience, investing in mobile-first marketing strategies, and enhancing app usability could further increase sales performance and customer engagement.

##### 5. Revenue by City

This section analyzes geographic performance by examining the number of orders and total revenue generated across different cities in Turkey. Understanding regional purchasing behavior helps identify key markets, assess revenue concentration, and inform strategic decisions related to marketing, logistics, and expansion planning.

<img src="Dashboard 1 copy.png" width="700" height="500" />

Key insights:

The results indicate that Istanbul generates the highest number of orders and the largest share of total revenue among all cities. This suggests that Istanbul represents the company’s strongest and most active market.

Business Insights:

Istanbul can be considered the company’s primary market, demonstrating both high customer activity and strong revenue contribution. Maintaining strong brand presence and customer satisfaction in this region should be a strategic priority.

The revenue distribution indicates geographic concentration in Istanbul, which may expose the company to regional dependency risk. Diversifying growth efforts in other cities could help balance revenue streams.

While Istanbul dominates performance, lower contributions from other cities suggest potential growth opportunities through targeted marketing campaigns, localized promotions, or improved regional logistics.

Conclusion:

Overall, the geographic analysis highlights Istanbul as the company’s most important revenue-driving market. While this strong performance reflects successful market penetration in the city, expanding customer engagement strategies in other regions may support more balanced and sustainable long-term growth.

#### 6. Average Rating by City

<img src="Average Rating by City.png" width="700" height="700" />

Business Insight:

2. Despite differences in revenue and order volume, customer satisfaction levels remain stable, suggesting effective operational processes and service management across all cities.

3. Istanbul’s strong revenue performance is not negatively impacting customer satisfaction, indicating that higher transaction volumes are being managed effectively without compromising service quality.

Conclusion:

Overall, the rating analysis suggests that the company delivers a consistent customer experience across its geographic markets. While revenue and order volumes vary significantly by city, customer satisfaction remains stable, reflecting strong operational performance.

# Final Overall Conclusion

5. Revenue Contribution by Customer Type
